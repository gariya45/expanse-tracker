# Exploratory Data Analysis - Personal Expense Tracker

This notebook performs exploratory data analysis on expense data to understand spending patterns, trends, and insights.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import sys
from pathlib import Path

# Add src directory to path
sys.path.insert(0, str(Path.cwd().parent))

# Import project modules
from src.data_loader import DataLoader
from src.preprocessing import DataPreprocessor
from src.categorization import ExpenseCategorizer
from src.analysis import ExpenseAnalyzer
from src.visualization import ExpenseVisualizer

# Set style
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

print("Libraries imported successfully!")
```

## 1. Load and Preprocess Data

```python
# Load data
loader = DataLoader()
df = loader.load_data()

print("Raw Data Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Preprocess data
preprocessor = DataPreprocessor()
df = preprocessor.preprocess_data(df)
df = preprocessor.add_time_features(df)

print("Preprocessed Data Shape:", df.shape)
print("\nData Info:")
df.info()
```

```python
# Check for missing values
print("Missing Values:")
df.isnull().sum()
```

## 3. Expense Analysis

```python
# Initialize analyzer
analyzer = ExpenseAnalyzer(df)

# Get insights
insights = analyzer.get_insights()

print("=== EXPENSE INSIGHTS ===")
print(f"Total Expenses: ${insights['total_expenses']:,.2f}")
print(f"Average Daily Spending: ${insights['average_daily']:,.2f}")
print(f"Average Monthly Spending: ${insights['average_monthly']:,.2f}")
print(f"Top Category: {insights['top_category']} (${insights['top_category_amount']:,.2f})")
print(f"Highest Spending Day: {insights['highest_spending_day']} (${insights['highest_spending_amount']:,.2f}")
print(f"Total Transactions: {insights['total_transactions']}")
```

## 9. Unnecessary Spending Analysis

```python
# Identify unnecessary spending
flagged = categorizer.identify_unnecessary_spending(df, threshold=100)

if len(flagged) > 0:
    print(f"Found {len(flagged)} potentially unnecessary expenses:")
    flagged
else:
    print("No unnecessary spending detected.")
```

## 10. Day of Week Analysis

```python
# Spending by day of week
day_of_week_spending = df.groupby('day_name')['amount'].agg(['sum', 'mean', 'count'])
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_of_week_spending = day_of_week_spending.reindex(day_order)
day_of_week_spending
```

```python
# Plot spending by day of week
fig, ax = plt.subplots(figsize=(10, 6))
day_of_week_spending['sum'].plot(kind='bar', color=sns.color_palette('viridis', 7))
ax.set_title('Total Spending by Day of Week', fontweight='bold')
ax.set_xlabel('Day of Week')
ax.set_ylabel('Total Amount ($)')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
